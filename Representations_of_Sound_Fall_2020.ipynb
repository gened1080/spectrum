# Representations of Sound

Sound can be represented graphically in two ways and are displayed below:

*  In a wave form whereby the signal is plotted as time against amplitude. 
*  As a spectrum of frequencies whereby the signal is plotted as frequency against amplitude. Frequency is often measured in Hz (oscillations per second) and as you can see below, the sine wave is one frequency and thus shows up as one "spike" in the spectrum representation.

Please run the code cells and feel free to play around with the frequency slider to see how the representations change.

We will be using these representations to help us understand Fourier Transforms and Fourier Series in the Spectrum Analysis notebook.

```python
def plot_wave_spec(Frequency):
    # create sine signal
    sig = thinkdsp.CosSignal(freq=Frequency, amp=0.6366, offset=0)

    # plot wave
    wave = sig.make_wave(duration=4.0, start=0, framerate=11025)
    segment = wave.segment(start=0, duration=0.01)
    segment.plot()
    thinkplot.config(xlabel='Time (s)')
    thinkplot.config(ylabel='Amplitude')
    thinkplot.config(title='Time domain representation')
    thinkplot.config(xlim=(0, 0.01))
    plt.show()

    # plot spec
    segment_spec = wave.make_spectrum()
    segment_spec.plot()
    thinkplot.config(xlabel='Frequency (oscillations/second)')
    thinkplot.config(ylabel='Amplitude')
    thinkplot.config(title='Frequency domain representation')
    thinkplot.config(xlim=(0,1100))
    plt.show()

    return None
```

```python
print('Move the slider to explore how sine waves with different frequencies (Hz) are represented')
interact(plot_wave_spec, Frequency=widgets.IntSlider(min=10, max=1000, step=20, value=440));
```
